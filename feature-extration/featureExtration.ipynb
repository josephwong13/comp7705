```python
# Loading testing data
testDataFile = open("testData.txt", "r")
spamSms = testDataFile.read().split("\n")
testDataFile.close()

print(len(spamSms))
print(spamSms[0])
```

```python
# Testing Bag of Word method
import sklearn
from sklearn.feature_extraction.text import CountVectorizer
vectorizer = CountVectorizer()
vectorizer.fit(spamSms)
vector = vectorizer.transform(spamSms)

print(vector.shape)
print(type(vector))
print(vector.toarray())
print(vectorizer.vocabulary_)
```

```python
# Transform word using bad of word vectorizer
firstSms = vectorizer.transform(spamSms[0].split('######'))

print(firstSms.shape)
print(type(firstSms))
print(firstSms)
```

```python
# Another test, the single word is ignored
anotherSms = vectorizer.transform("平安 平安 平安 天 天 天".split('######'))

print(anotherSms.shape)
print(anotherSms)
```

```python
# Testing tf-idf
from sklearn.feature_extraction.text import TfidfVectorizer

tfidFvectorizer = TfidfVectorizer()
tfidFvectorizer.fit(spamSms)

print(tfidFvectorizer.vocabulary_)
print(tfidFvectorizer.idf_)
```

```python
# encode document
tfidFvector = tfidFvectorizer.transform(spamSms[0].split('######'))
# summarize encoded vector
print(tfidFvector.shape)
print(tfidFvector)
print(tfidFvector.toarray())
```

```python
# Hashing vectorizer
from sklearn.feature_extraction.text import HashingVectorizer

hashingVectorizer = HashingVectorizer(n_features=5000)
hashVector = hashingVectorizer.transform(spamSms[0].split('######'))

print(hashVector.shape)
print(hashVector)
print(hashVector.toarray())
```
